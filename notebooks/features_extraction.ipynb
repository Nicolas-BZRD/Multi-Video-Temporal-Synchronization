```python
import os

import cv2 as cv
import numpy as np

from models.cnn import cnn
```

```python
mdl = cnn.FeauturesExtraction()
```

```python
numberImage = len(os.listdir(r"datasets\ISIA_cv_flow\1\left"))
```

```python
image = []

for i in range(numberImage):
    if i%1000==0: print(i)

    imgL = cv.resize(cv.imread(fr"datasets\ISIA_cv_flow\1\left\{i}.jpg"),(224,224))
    imgR = cv.resize(cv.imread(fr"datasets\ISIA_cv_flow\1\right\{i}.jpg"),(224,224))

    imgL = np.apply_along_axis(lambda x: x/255, 0, imgL)
    imgR = np.apply_along_axis(lambda x: x/255, 0, imgR)

    featuresL = mdl.extract(imgL)[0].tolist()
    featuresR = mdl.extract(imgR)[0].tolist()

    image.append([featuresL, featuresR, i])
```

```python
image = np.array(image)
np.save("isia_cv_flow_features", image)
```
