# Find delay with deep network

```python
import numpy as np

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import np_utils

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

## Data

```python
ds = np.load(r"..\datasets\isia_flow_similarity_with_delay.npy", allow_pickle=True)
```

```python
ds_test = np.load(r"..\datasets\click_flow_similarity_with_delay.npy", allow_pickle=True)
```

## Delay between -1 to 9

```python
tmp = ds.tolist()
data = []
for i in range(len(tmp)):
    if tmp[i][1]!=-1 and tmp[i][1]<10:
        data.append(tmp[i])

data = np.array(data, dtype=object)

X = data[:,0].tolist()
y = data[:,1].tolist()
```

```python
X_train = np.array(X)
y_train = np_utils.to_categorical(y, num_classes=10)
# X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.1, random_state=42, stratify=y)
```

```python
modelSmallDelay = Sequential([
    Dense(64, input_dim = 400, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.4),
    Dense(10, activation="softmax")
])

modelSmallDelay.compile(loss='categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
```

```python
modelSmallDelay.fit(X_train, y_train, epochs=500)
```

```python
tmp = ds_test.tolist()
data = []
for i in range(len(tmp)):
    if tmp[i][1]!=-1 and tmp[i][1]<10:
        data.append(tmp[i])

data = np.array(data, dtype=object)

X_test = data[:,0].tolist()
y_test = data[:,1].tolist()
```

```python
X_test = np.array(X_test)
y_test = np_utils.to_categorical(y_test, num_classes=10)
```

```python
modelSmallDelay.evaluate(X_test, y_test)
```

```python
y_pred_test_prob = modelSmallDelay.predict(X_test)
```

```python
y_pred_test = []
for i in range(len(y_pred_test_prob)):
    y_pred_test.append(np.argmax(y_pred_test_prob[i]))
```

```python
y_test_index = []
for i in range(len(y_test)):
    y_test_index.append(np.argmax(y_test[i]))
```

```python
print(classification_report(y_test_index, y_pred_test))
```

```python
print(confusion_matrix(y_true=y_test_index, y_pred=y_pred_test))
```

## Delay between 0 to 20

```python
data = ds.tolist()

for i in range(len(data)):
    if data[i][1] == 20: data[i][1] = -1

data = np.array(data, dtype=object)

X = data[:,0].tolist()
y = data[:,1].tolist()
```

```python
X_train = np.array(X)
y_train = np_utils.to_categorical(y, num_classes=21)
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)
```

```python
modelFullDelay.fit(X_train, y_train, epochs=400)
```

```python
data = ds_test.tolist()

for i in range(len(data)):
    if data[i][1] == 20: data[i][1] = -1

data = np.array(data, dtype=object)

X = data[:,0].tolist()
y = data[:,1].tolist()
```

```python
X_test = np.array(X)
y_test = np_utils.to_categorical(y, num_classes=21)
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)
```

```python
modelFullDelay.evaluate(X_test, y_test)
```

```python
y_pred_test_prob = modelFullDelay.predict(X_test)
```

```python
y_pred_test = []
for i in range(len(y_pred_test_prob)):
    y_pred_test.append(np.argmax(y_pred_test_prob[i]))
```

```python
y_test_index = []
for i in range(len(y_test)):
    y_test_index.append(np.argmax(y_test[i]))
```

```python
print(classification_report(y_test_index, y_pred_test))
```
